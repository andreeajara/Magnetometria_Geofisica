```python
mag=[]
for i in range(size(magr)):
    mag.append(magr[i]-magb[i])
amag = asarray(mag)
```

```python
scatter(vx,vy, linewidth=0.5, s=5, c='g')
plot(vx,vy, linewidth=0.15, c='g')
xlabel('x (m)')
ylabel('y (m)')
```

```python
data3 = loadtxt("grupoe2.txt", usecols = (0,1,2))
data4 = loadtxt("grupoe2base.txt", usecols = (0,1,2))
cmagr = data3[:,2]
cx=data3[:,0]
cy=data3[:,1]
cmagb=data4[:,1]
```

```python
mag2=[]
for i in range(size(cmagr)):
    mag2.append(cmagr[i]-cmagb[i])
camag = asarray(mag2)
```

```python
scatter(cx,cy, linewidth=0.1, s=1, c='g')
plot(cx,cy, linewidth=0.15, c='g')
xlabel('x (m)')
ylabel('y (m)')
```

```python
aa=[]
bb=[]
for i in range(size(cx)):
    aa.append(cx[i])
    bb.append(camag[i])
    
plot(aa,bb)
xlabel('x (m)')
ylabel('AIM (nT)')
```

```python
cz= camag
fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
ax.scatter(cx, cy, cz, linewidth=0, s=20, color='r')
ax.view_init(elev=0, azim=90)
plot(cx,cy, linewidth=0.15, c='g')
```
