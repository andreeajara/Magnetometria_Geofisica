```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = loadtxt("grupoa1.csv", delimiter=',', usecols = (0,1,2,9))
magr = data[:,2]
vx=data[:,0]
vy=data[:,1]
magb = data[:,3]
```

```python
mag = []
for i in range(size(magr)):
    mag.append(magr[i]-magb[i])
amag = asarray(mag)
```

```python
fig = plt.figure(figsize=(10,5))
scatter(vx,vy, linewidth=0.5, s=5, c='g')
plot(vx,vy, linewidth=0.5, c='g')
xlabel('x (m)')
ylabel('y (m)')
title('Rejilla en el plano xy del survey 1')
```

```python
vz = amag
fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
ax.scatter(vx, vy, vz, linewidth=0, s=50, color='r')
ax.view_init(elev=0, azim=0)
plot(vx,vy, linewidth=0.15, c='g')
```
